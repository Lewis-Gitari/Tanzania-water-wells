```python
# Get the number of rows for each unique value in the 'ward' column
rows_per_value = df['ward'].value_counts()

# Calculate the median number of rows
median_rows = rows_per_value.median()

print(f"Median number of rows: {median_rows}")
```

```python
import numpy as np

# Set a seed for reproducibility
np.random.seed(42)

# Get the number of rows for each unique value in the 'ward' column
rows_per_value = df['ward'].value_counts()

# Calculate the median number of rows
median_rows = rows_per_value.median()

# Round the median value to the nearest whole number just incase it's a float.
rounded_median_rows = int(round(median_rows))

# Create an empty DataFrame to store the final result
final_df = pd.DataFrame()

# Iterate over unique values in the 'ward' column
for unique_value in df['ward'].unique():
    # Get the rows for the current unique value
    rows_for_value = df[df['ward'] == unique_value]
    
    # Check if there are more rows than the rounded median
    if len(rows_for_value) > rounded_median_rows:
        # Randomly choose rows to drop
        rows_to_drop = rows_for_value.sample(n=len(rows_for_value) - rounded_median_rows)
        
        # Drop the randomly chosen rows
        rows_for_value = rows_for_value.drop(rows_to_drop.index)
    
    # Append the rows to the final DataFrame
    final_df = pd.concat([final_df, rows_for_value])

print(final_df.head())
```
